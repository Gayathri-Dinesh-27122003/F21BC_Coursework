# PSO Example: Optimize Mathematical Function

Simple demonstration of Particle Swarm Optimization on a 2D Sphere function with convergence visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pso import PSO

print("✅ Libraries imported")
```

## Define Objective Function

Optimize the Sphere function: f(x) = x₁² + x₂²

```python
def sphere_function(x):
    """
    Sphere function: f(x) = sum(x_i^2)
    Global minimum at (0, 0) with value 0
    """
    return np.sum(x**2)
```

## Run PSO Optimization

```python
# Initialize PSO optimizer
pso = PSO(
    objective_function=sphere_function,
    dimension=2,               # Optimize 2D problem
    swarm_size=20,            # 20 particles
    max_iterations=100,       # 100 iterations
    bounds=(-5.12, 5.12)      # Search space [-5.12, 5.12]^2
)

# Run optimization
best_params, best_fitness, history = pso.optimize(verbose=False)

# Display results
print("=" * 50)
print("PSO Optimization Results")
print("=" * 50)
print(f"Best Solution Found: {best_params}")
print(f"Best Fitness Value: {best_fitness:.6f}")
print(f"Expected Optimum: [0, 0]")
print(f"Expected Value: 0.0")
print("=" * 50)
```

## Convergence Visualization

```python
plt.figure(figsize=(10, 6))
plt.plot(history, 'b-', linewidth=2, label='Best Fitness')
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Best Fitness Value', fontsize=12)
plt.title('PSO Convergence on Sphere Function', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

print(f"Convergence achieved after ~{len(history)} iterations")
print(f"Final improvement over initial: {(history[0] - history[-1]) / history[0] * 100:.1f}%")
```
